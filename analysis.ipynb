add new column using np where condition

```python

import numpy as np


# Ensure 'born_date' and 'died_date' are datetime objects
updated_athlete_df['born_date'] = pd.to_datetime(updated_athlete_df['born_date'])
updated_athlete_df['died_date'] = pd.to_datetime(updated_athlete_df['died_date'], errors='coerce') # If 'coerce' , then invalid parsing will be set as NaN

# Calculate 'age' in years for living athletes
updated_athlete_df['age'] = np.where(
    pd.notnull(updated_athlete_df['died_date']),
    (updated_athlete_df['died_date'] - updated_athlete_df['born_date']).dt.days / 365.25,
    (pd.to_datetime('today') - updated_athlete_df['born_date']).dt.days / 365.25
)

# Create 'status' column indicating deceased status
updated_athlete_df['status'] = np.where(
    pd.notnull(updated_athlete_df['died_date']),
    'Dead',
    'Alive'
)
display(updated_athlete_df)
```

str.split() and inserting the new column in desired location

```python
# rename the column name
updated_athlete_df.rename(columns={'name':'full_name'},inplace = True)

# Split 'name' into 'first_name' and 'last_name'
updated_athlete_df[['first_name', 'last_name']] = updated_athlete_df['full_name'].str.split(' ', n=1, expand=True)

# Insert 'first_name' at column 2 (index 2) and 'last_name' at column 3 (index 3)
updated_athlete_df.insert(2, 'first_name', updated_athlete_df.pop('first_name'))
updated_athlete_df.insert(3, 'last_name', updated_athlete_df.pop('last_name'))

# Display the DataFrame to verify the result
updated_athlete_df
```

Using Lambda Function and inserting the column in desired location using insert and apply

```python
updated_athlete_df['height_category'] = updated_athlete_df['height_cm'].apply(
    lambda x: 'Short' if x < 145 else ('Average' if x < 185 else 'Tall')
)

# Insert the 'height_category' column at position 2 (index 2)
updated_athlete_df.insert(10, 'height_category', updated_athlete_df.pop('height_category'))

# Display the updated DataFrame
updated_athlete_df
```

```python
def weight_function(bio):
    if bio['height_cm']<175  and bio['weight_kg']<70:
        return 'Lightweight'
    elif bio['height_cm']<185 or bio['weight_kg']<=80:
        return 'Heavyweight'

updated_athlete_df['weight_category']=updated_athlete_df.apply(weight_function, axis = 1)
updated_athlete_df.insert(12,'weight_category',updated_athlete_df.pop('weight_category'))
updated_athlete_df
```

Using Interpolation to manage na


```python
no_na_athlete_df = updated_athlete_df.copy()
no_na_athlete_df[['height_cm','weight_kg']] = no_na_athlete_df[['height_cm','weight_kg']].interpolate() ## interpolate estimates the value of a function at points where the value is not known, based on given discrete data points
no_na_athlete_df.iloc[145152:145495]
```
